```python
%matplotlib inline
import matplotlib.pyplot as plt
from mwviews.api import PageviewsClient
```

```python
p = PageviewsClient()
page = 'Deep_learning'
response = p.article_views('en.wikipedia', [page], start='20160101', end='20170201')
```

```python
dates = response.keys()
pageviews = []
for date in dates:
    if date != 0:
        elem = response[date]
        pv = elem[page]
        pageviews.append(pv)
```

```python
plt.plot(pageviews)
plt.show()
```
